```python
import pandas as pd
```

```python
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'
```

```python
column_names = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'class']
```

```python
df = pd.read_csv(url, header=None, names=column_names)

df.head()
```

## 1. List down the features and their types (e.g., numeric, nominal) available in the dataset

Looking at the Iris dataset, we can identify the following features:

- Sepal Length (numeric)
- Sepal Width (numeric)
- Petal Length (numeric)
- Petal Width (numeric)
- Class (nominal)

The "Class" column represents the target variable, specifying the iris species for each sample.

## 2. Create a histogram for each feature in the dataset to illustrate the feature distributions

```python
import matplotlib.pyplot as plt
```

```python
# create hostogram for each feature

# figsize parameter determines the size of the figure
df.hist(figsize=(10, 6))

# tight_layout() ensures proper spacing between the subplots
plt.tight_layout()

# plt.show()
```

## 3. Create a boxplot for each feature in the dataset

```python
df.boxplot(figsize=(10, 6))

plt.tight_layout()
```

## 4. Compare distributions and identify outliers

By examining the histograms and boxplots, we can compare the distributions of each feature and identify any potential outliers. Histograms provide a visual representation of the data distribution, while boxplots help identify outliers based on the presence of points beyond the whiskers.

In the histograms, pay attention to any significant deviations from a normal distribution and note the range and concentration of values in each feature. In the boxplots, look for individual points plotted beyond the whiskers, as they may indicate potential outliers.

Analyzing the histograms and boxplots, you can make observations about the data distribution and identify any potential outliers that lie outside the whiskers in the boxplots.
